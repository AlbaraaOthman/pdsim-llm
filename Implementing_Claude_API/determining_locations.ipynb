```python
with open("../plans/bricks.txt", "r") as file:
    content = file.read()
    # print(content)
    fixed = content.replace('[[[', '[').replace('[[','[').replace(']]]',']').replace(']]',']')
    # print(fixed)
    fixed_array = ast.literal_eval(fixed)
    for array in fixed_array:
        print(array)
```
